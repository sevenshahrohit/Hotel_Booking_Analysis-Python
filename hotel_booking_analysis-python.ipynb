# data analysis & visualisation

```python
cancel_per = hba['is_canceled'].value_counts(normalize=True)
cancel_per
#this will give the % counts of cancelation
#0 is not cancel & 1 is cancel
```

```python
plt.bar(['Not Cancel','Cancel'],hba['is_canceled'].value_counts(),color=['green','red'])
plt.title('Reservation Status count')
plt.show()
```

```python
sns.countplot(data=hba ,x='hotel', hue='is_canceled')
plt.title('Reservation status in different Hotels')
plt.legend(['Not Cancel','Cancel'])
plt.ylabel('No. of Reservation')
plt.xlabel('Hotels')
plt.show()
```

```python
resort_hotel = hba[hba['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
#this will show us the resort hotel cancelation %
#0 is no cancel
#1 is cancel
```

```python
city_hotel = hba[hba['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
#this will show us the city hotel cancelation %
```

```python
hba['month']=hba['reservation_status_date'].dt.month
sns.countplot(data=hba ,x='month', hue='is_canceled')
plt.title('Reservation status per Month')
plt.legend(['Not Cancel','Cancel'])
plt.ylabel('No. of Reservation')
plt.xlabel('month')
plt.show()
```

```python
cancelled_data = hba[hba['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts(normalize=True)[:10]
#this will give top 10 country with most number of cancelations
```

```python
plt.pie(top_10_country,labels=top_10_country.index,autopct='%.01f%%')
plt.show()
#autopct='%0.2f%%': yh values show krne k liye use krte hai
```

```python
hba['market_segment'].value_counts()
```

```python
hba['market_segment'].value_counts(normalize=True)
```

```python
 cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
cancel_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancel_adr.reset_index(inplace=True)
cancel_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data=hba[hba['is_canceled']==0]
not_cancel_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancel_adr.reset_index(inplace=True)
not_cancel_adr.sort_values('reservation_status_date',inplace=True)

plt.title('ADR')
plt.plot(not_cancel_adr['reservation_status_date'],not_cancel_adr['adr'],label='Not Cancel')
plt.plot(cancel_adr['reservation_status_date'],cancel_adr['adr'],label='Cancel')
plt.legend(loc='upper right')
plt.show()
```
